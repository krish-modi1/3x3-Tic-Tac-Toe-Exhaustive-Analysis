### Q2 Summary (24 Points):

Summarize the main objectives and contributions of the paper.

A2:

Main Objective: Implementing and fabricating a CNN accelerator called Eyeriss that supports high throughput CNN inference and optimizes for the energy efficiency of the entire system, including the accelerator chip and off-chip DRAM

Its main contributions are:

1. A spatial architecture using an array of 168 processing elements that creates a four-level memory hierarchy to minimize data accesses to high-cost levels by exploiting data movements at low-cost levels
2. A CNN dataflow called Row Stationary (RS) that maximises data reuse and minimises data movement energy
3. A network-on-chip (NoC) architecture using multicast and point-to-point single-cycle data delivery supporting RS dataflow
4. Integrating run-length compression (RLC) and data gating to further improve energy efficiency for CNNs

The performance of Eyeriss is benchmarked against the two widely used and public CNNs, AlexNet and VGG-16

### Q4 Technical Deep Dive (15 Points):
- Describe the spatial architecture of Eyeriss in your own words.
- How does the Eyeriss architecture optimize energy efficiency in dataflow for CNNs?
- What are the main challenges in designing an energy-efficient architecture for CNNs, and how does Eyeriss tackle these challenges?

A4:

- Spatial architecture of Eyeriss:
  1. Eyeriss uses a 12x14 2D grid of 168 identical processing elements. Each PE has its local scratchpad memories, forming a 4-level memory hierarchy:
    1. DRAM
    2. Global buffer
    3. inter-PE communication
    4. PE scratchpads

- Eyeriss optimises energy efficiency in dataflow using Row Stationary dataflow where PEs are grouped into sets that implement 1D and 2D convolution primitives. Hence, nearby PEs share rows of filters, input feature maps, and partial sums via local links and the custom NoC instead of accessing the buffer or DRAM constantly. Further improvements come from using run length compression for feature maps and zero skipping/data-gating in PEs

- The main challenges are that CNNs require large amounts of data movements on or off chip, which are energy-expensive compared to MAC computations, while the hardware must stay energy-efficient and highly parallel for any CNN layer shapes. Eyeriss tackles it by using the spatial architecture explained above

### Q5 Evaluation (15 Points):
- What were the key results or findings of the paper? Were they compelling?
- How do the authors validate their claims or results? Are there any weaknesses in their methodology?

- Eyeriss achieves high energy efficiency and low DRAM traffic while running AlexNet and VGG-16:
  1. For AlexNet, it reaches 35 frames/s at 278mW with only 0.0029 DRAM accesses per MAC
  2. For VGG-16, it reaches 0.7 frames/s at 236 mW with 0.0035 DRAM access per MAC
  
  These are compelling results because they are measured on a 65 nm chip, benchmarked on widely used public CNNs, including DRAM bandwidth numbers, and they confirm that most power is in data movement and not ALUs, showing that RS dataflow and memory-centric design actually improve overall energy efficiency

- The authors validate their claims by measuring real performance markers like throughput, power, energy efficiency, DRAM accesses on public and widely used CNNs, with power breakdown showing most energy is used by data movement

  A potential weakness can be that the results are only for two CNNs and a single 65 nm technology node, so it might not be fully generalised for other workloads. Also, there is no head-to-head comparison against other accelerators in these conditions

### A6:

  The reconfigurability of Eyeriss for different CNN shapes also points toward a more general design philosophy where accelerators adapt to the varying layer dimensions found in modern deep networks rather than being hardwired for a single configuration.

- The following principles of Eyeriss might be applied to other deep learning architectures beyond CNNs:
